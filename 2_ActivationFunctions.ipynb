```python
import math

def sigmoid(x):
  return 1/(1+ math.exp(-x))
```

```python
print(sigmoid(100))
print(sigmoid(1))
print(sigmoid(-56))
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
print(tanh(100))
print(tanh(1))
print(tanh(-56))
```

```python
def relu(x):
  return max(0,x)
```

```python
print(relu(100))
print(relu(1))
print(relu(-56))
```

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
print(leaky_relu(100))
print(leaky_relu(1))
print(leaky_relu(-56))
```
