```python
import numpy as np

units_sold = np.array([
    [50, 60, 25],
    [10, 13, 5],
    [40, 70, 52]
])

price_per_unit = np.array([20, 30, 15])

total_sales = price_per_unit @ units_sold
total_sales
```
